# Mapa de China (2da división administrativa)

```python
import matplotlib.pyplot as plt

China.plot(facecolor="lightgreen", 
               edgecolor='blue',       
               linewidth=0.1)       
plt.title('Mapa de China con provincias', fontsize=15)
plt.savefig('mapa_china_provincias.png', dpi=300, bbox_inches='tight')
```

```python
rivers=gpd.read_file(os.path.join("data","World_Hydrography","World_Hydrography.shp"))
cities=gpd.read_file(os.path.join("data","World_Cities","World_Cities.shp"))
```

```python
China.crs==rivers.crs==rivers.crs
```

```python
riversChina_clipped = gpd.clip(gdf=rivers,
                               mask=China)
citiesChina_clipped = gpd.clip(gdf=cities,
                          mask=China)
```

# Mapa con divisiones geograficas y rios

```python
base = China.plot(facecolor="greenyellow", edgecolor='black', linewidth=0.4,figsize=(5,5))
riversChina_clipped.plot(marker='+', color='blue', markersize=15,
                    ax=base)
plt.title('Mapa de China con sus rios y provincias', fontsize=15)
plt.savefig('mapa_china_rios1.png', dpi=300, bbox_inches='tight')
```

```python
China_map =gpd.read_file(os.path.join("data","CHN_shp","CHN_0.shp"))
```

# China y sus rios

```python
base = China_map.plot(facecolor="greenyellow", edgecolor='black', linewidth=0.4,figsize=(5,5))
riversChina_clipped.plot(marker='+', color='blue', markersize=15,
                    ax=base)
plt.title('Mapa de China solo con sus rios', fontsize=15)
plt.savefig('mapa_china_rios2.png', dpi=300, bbox_inches='tight')
```

# Las ciudades de China en el mapa 

```python
base = China_map.plot(facecolor="greenyellow", edgecolor='black', linewidth=0.4,figsize=(5,5))
citiesChina_clipped.plot(marker='+', color='red', markersize=15,
                    ax=base)
plt.title('Ciudades de China', fontsize=15)
plt.savefig('mapa_china_ciudades.png', dpi=300, bbox_inches='tight')
```

# Mapa con divisiones geograficas, ciudades y rios 

```python
base = China.plot(facecolor="greenyellow", edgecolor='black', linewidth=0.4,figsize=(5,5))
riversChina_clipped.plot(marker='+', color='blue', markersize=15,
                    ax=base)
citiesChina_clipped.plot(marker='+', color='red', markersize=15,
                    ax=base)
plt.title('Mapa con divisiones geograficas, ciudades y rios ', fontsize=15)
plt.savefig('mapa_china_divisiones_ciudades_rios.png', dpi=300, bbox_inches='tight')
```

# Spatial Data

```python
import pandas as pd
infoairports=pd.read_csv(os.path.join("data","cn-airports.csv"))

# some rows

infoairports.iloc[[0,1,2,3,-4,-3,-2,-1],:] #head and tail
```

```python
infoairports.drop(index=0,inplace=True)
infoairports.reset_index(drop=True, inplace=True)
infoairports.head()
```

```python
infoairports.columns
```

```python
keep=['name','type','latitude_deg', 'longitude_deg','elevation_ft','region_name','municipality']
infoairports=infoairports.loc[:,keep]
```

```python
infoairports.info()
```

```python
numericCols=['latitude_deg', 'longitude_deg','elevation_ft']
infoairports[numericCols]=infoairports.loc[:,numericCols].apply(lambda x:pd.to_numeric(x))

# now
infoairports.info()
```

```python
airports=gpd.GeoDataFrame(data=infoairports.copy(),
                 geometry=gpd.points_from_xy(infoairports.longitude_deg,
                                             infoairports.latitude_deg),
                 crs=China.crs.to_epsg())# the coordinates were in degrees - unprojected
```

```python
base = China_5641.plot(color='white', edgecolor='black')
airports.plot(ax=base)
```

# Mapa de los aeropuertos en China

```python
airports_5641=airports.to_crs(5641)

## then

base = China_5641.plot(color='white', edgecolor='black')
airports_5641.plot(ax=base)
plt.title('Aeropuertos en China por provincias', fontsize=15)
plt.savefig('mapa_china_aeropuertos_prov.png', dpi=300, bbox_inches='tight')
```

```python
base = China_map_5641.plot(color='white', edgecolor='black')
airports.plot(ax=base)
```

```python
airports_5641=airports.to_crs(5641)

## then

base = China_map_5641.plot(color='white', edgecolor='black')
airports_5641.plot(ax=base)

plt.title('Aeropuertos en China', fontsize=15)
plt.savefig('mapa_china_aeropuertos.png', dpi=300, bbox_inches='tight')
```
